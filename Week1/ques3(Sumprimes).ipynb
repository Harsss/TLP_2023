### Write a function sumprimes(l) that takes as input a list of integers l and retuns the sum of all the prime numbers in l.

### Here are some examples to show how your function should work.
####  >>> sumprimes([3,3,1,13])
####  19
####  >>> sumprimes([2,4,6,9,11])
####  13
####  >>> sumprimes([-3,1,6])
####  0

#### 1. Harsh Approach
Make a list of prime number.then iterate on list and checking if the number in given list is divisible by numbers on prime list, if it is then remove it from  list else start checking its divisibility from last number of prime list to that number by last number and if found then delete from the list Else continue and at last add all the remaining number from the list  

```python
def sumprimes(ls):
    x=[2,3,5,7,11]
    v=ls.copy()
    n=len(ls)
    total=0
    for i in range(0,n):
        c=ls[i]
        if c in x:
            continue
        elif c==1:
            v.remove(c)
        else:
            for k in range(0, len(x)):
                if c%x[k]==0:
                    v.remove(c)
                    break
                else:
                    pass
            if c>120 or k!=4:
                continue
            else:
                for k in range (12,int(c/11)+1):   
                    if c%k==0:
                        v.remove(c)
                        break
                 
    for i in range(0,len(v)):
        total=total+v[i]
#     print(v)
    return(total)
```
