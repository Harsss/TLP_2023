### Write a function intreverse(n) that takes as input a positive integer n and returns the integer obtained by reversing the digits in n.
### Here are some examples of how your function should work.
###  >>> intreverse(783)
###  387
###  >>> intreverse(242789)
###  987242
###  >>> intreverse(3)
###  3

#### 1. Harsh's Approach 
By simple string slices, start from Last and go one step backward  at every step

```python
def intreverse(a):
    if(type(a)==str):
        a=a[-1::-1]
        return a
    else:
        x=type(a)
        a=str(a)
        a=a[-1::-1]
        return x(a)
```
