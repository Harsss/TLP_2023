### Write a function matched(s) that takes as input a string s and checks if the brackets "(" and ")" in s are matched: that is, every "(" has a matching ")" after it and every ")" has a matching "(" before it. Your function should ignore all other symbols that appear in s. Your function should return True if s has matched brackets and False if it does not.

### Here are some examples to show how your function should work.

### matched("zb%78")
### True
### >>> matched("(7)(a")
### False
### >>> matched("a)*(?")
### False
### >>> matched("((jkl)78(A)&l(8(dd(FJI:),):)?)")
### True


#### 1. Harsh Approach 
Start with an initial flag=0 and increase it with every opening bracket and decrease on every closing bracket and if at last it will be 0 withour being negative at any point then its True else false

```python
def matched(a):
    n=len(a)
    flag=0
    for i in range(0,n):
        if flag<0:
            return False
        elif a[i]=="(":
            flag=flag+1
        elif a[i]==")":
            flag=flag-1
        else:
            pass
        
    if flag==0:
        return True
    else:
        return False
```
